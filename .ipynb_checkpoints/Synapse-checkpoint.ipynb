```python
patient_dict = {}

for index, row in all_forms.iterrows():
    if row['SubjectID'] in patient_dict:
        if row['form_name'] == 'Concomitant Medication':
            patient_dict[row['SubjectID']]['Concomitant'] += 1
        elif row['form_name'] == 'Adverse Event':
            patient_dict[row['SubjectID']]['Adverse'] += 1
        else:
            patient_dict[row['SubjectID']][row['form_name']+'-'+row['feature_name']] = row['feature_value']
    else:
        patient_dict[row['SubjectID']] = {}
        patient_dict[row['SubjectID']]['Concomitant'] = 0
        patient_dict[row['SubjectID']]['Adverse'] = 0
        if row['form_name'] == 'Concomitant Medication':
            patient_dict[row['SubjectID']]['Concomitant'] += 1
        elif row['form_name'] == 'Adverse Event':
            patient_dict[row['SubjectID']]['Adverse'] += 1
        else:
            patient_dict[row['SubjectID']][row['form_name']+'-'+row['feature_name']] = row['feature_value']
```

```python
import os
import sys

dAs = {}
path = 'data/Synapse/no_onehot/'
for file in os.listdir(path):
    if file.endswith(".p"):
        d = pkl.load(open(path + '/' + file, 'rb'))
        nodes = file.split('.')[0]
        dAs[nodes] = d
```

```python
ev = {}
rms = {}
pearson = {}

cv = KFold(y.shape[0], n_folds=10, shuffle=True, random_state=1)
# regressions should be fit via CV
methods = {'svm': svm, 'ridge':ridge, 'lasso':lasso, 'lr':lr, 'rfc':rfc}

for method in methods:
    e = []
    r = []
    p = []
    for i, (train, test) in enumerate(cv):
        predictions = methods[method].predict(X[train], X[test], y[train])
        indi_scores = evaluator.evaluate(y[test], predictions)
        e.append(indi_scores[0])
        p.append(indi_scores[1]) 
        r.append(indi_scores[2])
           
    ev[method] = e
    pearson[method] = p
    rms[method] = r
print(rms)
```

```python
import theano
import theano.tensor as T
from theano.tensor.shared_randomstreams import RandomStreams

for nodes, d in dAs.items():
    i_theano = T.dmatrix('i_theano')
    get_hidden = d.get_hidden_values(i_theano)
    f = theano.function([i_theano], [get_hidden])
    x_hidden = f(X)[0]
    for method in methods:
        e = []
        r = []
        p = []
        for i, (train, test) in enumerate(cv):
            predictions = methods[method].predict(x_hidden[train], x_hidden[test], y[train])
            indi_scores = evaluator.evaluate(y[test], predictions)
            e.append(indi_scores[0]) 
            p.append(indi_scores[1])
            r.append(indi_scores[2])
        ev[str(nodes) + '_' + method] = e
        pearson[str(nodes) + '_' + method] = p
        rms[str(nodes) + '_' + method] = r
    print(nodes)
```
