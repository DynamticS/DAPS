```python
import os
import sys

import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd

%matplotlib inline 

dataset = 'ex2_1'
path = '../data/' + dataset + '/scores.txt'
with open(path, 'r') as f:
    scores_dict = eval('[' + f.read() + ']')[0]
            
# print(scores_dict)

for key, val in scores_dict.items():
    plt.figure(figsize=(20,10))
    x1,x2,y1,y2 = plt.axis()
    plt.axis((x1,x2,0,1))
    df = pd.DataFrame.from_dict(val, orient='columns', dtype=None)
    print(df.columns)
#     df.drop(['da_2_snn', 'da_2nn', 'da_4_snn', 'da_4nn', 'da_9_snn', 
#              'da_9nn', 'da_9_sfullrfc', 'da_9_srfc', 'da_9_ssvm', 'da_9fullrfc',
#              'da_9rfc', 'da_9svm', 'nn'], axis=1, inplace=True)
#     df.drop(['da_2nn', 'da_4nn', 'da_9nn', 'nn'], axis=1, inplace=True)
    sns.set_style("whitegrid")
    sns.set_context("poster")
    g = sns.boxplot(data=df, palette="husl")
    g.set_title('Effects: ' + str(key))
```
